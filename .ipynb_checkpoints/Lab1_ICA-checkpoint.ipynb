```python
#Functions needed
%pylab inline
# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
 return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp
def sine_wave(x, period=0.2, amp=1.0, phase=0.):
 return np.sin((x / period - phase) * 2 * np.pi) * amp
def square_wave(x, period=0.2, amp=1.0, phase=0.):
 return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) 
def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp
def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X,title = " "):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
        if i==0:
            plt.title(title)
```

```python
# Generate sample data
num_sources = 5
signal_length = 500
t = linspace(0, 1, signal_length)
S = np.c_[sawtooth(t), sine_wave(t, 0.3), square_wave(t, 0.4), triangle_wave(t, 0.25), np.random.randn(t.size)].T

plot_signals(S,"Plot of the original sources")
```

### 1.1 Make Mixtures

```python
def make_mixtures(S, A):
    X = np.dot(A,S)
    return X
#Get the Mixing Matrix
A = random_nonsingular_matrix(d=5)
#Get the mixed signals X
X = make_mixtures(S,A)
#Plot the mixed signals X
plot_signals(X,"Plot of the mixed measurements")
```

### 1.2 Histogram

```python
def plot_histograms(data,title):
    #histograms = [[] for i in range(len(X))]
    for row in xrange(len(data)):
        hist, bins = np.histogram(data[row,:],bins = 30)
        width = 0.7 * (bins[1] - bins[0])
        center = (bins[:-1] + bins[1:]) / 2
        plt.xlabel("bins")
        plt.ylabel("frequency")
        plt.title(title)
        plt.bar(center, hist, align='center', width=width)
        plt.show()
rcParams['figure.figsize'] = 6, 5
#Plot Histograms of the Source signals
plot_histograms(S,"Histogram of source")
#Plot Histograms of the Mixed signals
plot_histograms(X,"Histogram of mixed measurement")
```

As can be seen above the mixed measurements clearly tend to look more like Gaussians. This can be explained by the central limit theorem which states that, given certain conditions, the arithmetic mean of a sufficiently large number of iterates of independent random variables, each with a well-defined expected value and well-defined variance, will be approximately normally distributed (Wikipedia). So the distribution of a sum of independent signals with arbitrary distributions tends toward a Gaussian distribution. By multiplying the source signals with a random mixing matrix we have generated random indepedent observations that follow this theorem. Thus a gaussian signal can be considered as a linear combination of many independent signals. This is important for ICA because the separation of independent signals from their mixtures is based to the fact that the latent variables (sources) are as non-Gaussian as possible. The Gaussian distribution of the latent variables is invariant under rotation of the latent variables, so there can be no evidence favouring any particular allignment of the latent variable space 

### 1.3 Implicit priors

Derivation of source distributions: 

$$\bullet \phi_0(a) = -\tanh(a) \implies \phi_0(a) = \frac{d}{da} \ln p(a) \implies -\tanh(a) = \frac{d}{da} \ln p(a) \implies \int -\tanh(a) da = \ln p(a) \implies -\int \frac{sinh(a)}{cosh(a)}da = \ln p(a) \\ \implies -\ln (cosh(a))+C = \ln p(a).\\$$ The answer is $$ p(a) \propto  \frac{1}{cosh(a)}$$

---

$$\bullet  \phi_1(a) = -a + \tanh(a) \implies \phi_1(a) = \frac{d}{da} \ln p(a) \implies -a + \tanh(a) = \frac{d}{da} \ln p(a) \implies \int -a + \tanh(a) da = \ln p(a) \implies \int tanh(a) da - \int a da = \ln p(a) \\  \implies \int tanh(a)da - \frac{a^2}{2} = \ln p(a) \implies \int \frac{sinh (a)}{cosh (a)} da - \frac{a^2}{2} = \ln p(a) \implies \ln cosh(a) -\frac{a^2}{2} + C = \ln p(a)\\$$ The answer is $$ p(a) \propto e^\frac{-a^2}{2} cosh(a)$$

---

$$\bullet  \phi_2(a) = -a^3 \implies \phi_2(a) = \frac{d}{da} \ln p(a) \implies -a^3 = \frac{d}{da} \ln p(a) \implies \int -a^3 da = \ln p(a) \implies -\frac{a^4}{4}+C = \ln p(a) \\$$ The answer is $$ p(a) \propto e^\frac{-a^4}{4}$$

---

$$\bullet  \phi_3(a) = -\frac{6a}{a^2 + 5} \implies \phi_3(a) = \frac{d}{da} \ln p(a) \implies -\frac{6a}{a^2 + 5} = \frac{d}{da} \ln p(a) \implies \int -\frac{6a}{a^2 + 5} = \ln p(a) \implies -6 \int \frac{a}{a^2 + 5} = \ln p(a) \\
u = a^2 + 5 \implies du = 2 a da \\$$ So we have: $$-3 \int \frac{1}{u} = -3 \ln (u) \\$$ Finally $$\ln p(a) = -3 \ln (a^2 +5)+C \\$$ The answer is $$ p(a) \propto \frac{1}{(a^2 +5)^3}$$

### 1.4 Whitening

```python
def whiten(X):

   from numpy import dot, sqrt, diag
   from numpy.linalg import eigh

   # the matrix X should be observations-by-components

   # get the covariance matrix
   Xcov = np.cov(X)

   # eigenvalue decomposition of the covariance matrix
   d,V = eigh(Xcov)

   # a fudge factor can be used so that eigenvectors associated with
   # small eigenvalues do not get overamplified.
   D = diag(1./sqrt(d))

   # whitening matrix
   W = dot(dot(V,D),V.T)

   # multiply by the whitening matrix
   Xwhiten = dot(W,X)

   return Xwhiten

Xwhiten = whiten(X)
```

### 1.5 Interpret results of whitening 

```python
# scatter plots of the sources(left), measurements(middle) and whitened measurements(right)
def scatter_plot(S,X,Xwhiten):
    count = 1
    figure()
    for i in range (0,4):
        for j in range (i+1,5):
            subplot(10,3,count)
            title("Source")
            plt.scatter(S[i],S[j],s=0.9,c='b')
            subplot(10,3,count+1)
            title("data")
            plt.scatter(X[i],X[j],s=0.9,c='r')
            subplot(10,3,count+2)
            title("Whitened data")
            plt.scatter(Xwhiten[i],Xwhiten[j],s=0.9,c='g')
            count += 3
rcParams['figure.figsize'] = 16, 40
scatter_plot(S,X,Xwhiten)
```

As can be seen fron the above figure the plots of the data indicate tha the mixed signals of the measurements are correlated. Whitening is an intermediate step before ICA is applied and it is used to uncorrelate the data. This reduces ICA to finding an appropriate rotation to yield indepedence (demixing the signals).

## 1.7 Independent Component Analysis

```python
from numpy import linalg as la

#activation functions
a0 = lambda x: -np.tanh(x);
a1 = lambda x: -x + tanh(x);
a2 = lambda x: -x**3;
a3 = lambda x: -((6*x)/(x**2 + 5));

##############################Online ICA##################################
def onlineICA(X, activation_function,iterations, lr=0.001):
    #Whiten the data
    X = whiten(X)
    #Create random unmixing matrix
    W = random_nonsingular_matrix(len(X))
    for m in range(0,iterations):
        for x in X.T:
            a = dot(W,x)
            z = activation_function(a)
            xPrime = dot(W.T,a)
            #Wpinv = linalg.inv(W.T)
            dW = zeros((W.shape[0],W.shape[0]))
            for i in range(W.shape[0]):
                for j in range(W.shape[0]):
                    dW[i][j] = (W[i][j] + xPrime[j]*z[i])
            W = W + lr*dW
    anew = dot(W,X)
    return anew

############################ICA in batch mode##############################
def batchICA(X, activation_function, lr = 0.01 , thresh = 0.01):
    #Whiten the data
    X = whiten(X)
    #Create random unmixing matrix
    W = random_nonsingular_matrix(len(X))
    count = 0;
    check = 2;
    while check > thresh:
        dW = zeros((W.shape[0],W.shape[0])) #create a matrix of size W with zeros
        a = dot(W,X)                        #Put x through a linear mapping
        z = activation_function(a)          #Put a through a nonlinear map:
        xPrime = dot(W.T,a)                 #Put a back through W
        dW = W*X.shape[1] + dot(z,xPrime.T) #Adjust the weights
        W = W + lr*(dW/X.shape[1])          #Update the weights
        check = sum(np.absolute(dW/X.shape[1]))
        count +=1
        #print iterations and check for convergence
        if count%100 == 0 : print count, check
    anew = dot(W,X)
    return anew
```

```python
#########################Plot the demixed signals#######################
A0 = batchICA(X,a0)
A1 = batchICA(X,a1)
A2 = batchICA(X,a2)
A3 = batchICA(X,a3)

rcParams['figure.figsize'] = 7, 5
plot_signals(A0,"Plot of the demixed signals using activation function a0")
plot_signals(A1,"Plot of the demixed signals using activation function a1")
plot_signals(A2,"Plot of the demixed signals using activation function a2")
plot_signals(A3,"Plot of the demixed signals using activation function a3")
```

We can see that by using activation functions a1 and a2 we get pretty good results as the derived signals from ICA matches the source signals.

## 1.9 Excess Kurtosis

### Analytical Kurtosis computation
All the pdfs we are studying have $\mu = 0$ so the kurtosis can be calculated as   $\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3$ . The definite integrals needed for calculating the expectations were calculated using Wolfram.

---

$$p(a) \propto \frac{1}{cosh(a)}$$
The normalisation constant as calculated by scipy.integrate is $C = \frac{1}{\pi}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{1}{\pi} \int\limits_{-\infty}^{\infty} \frac{1}{cosh(a)} a^4 da = \frac{1}{\pi}  \frac{5 \pi^5}{16}  =  \frac{5 \pi^4}{16}
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{1}{\pi} \int\limits_{-\infty}^{\infty} \frac{1}{cosh(a)} a^2 da =  \frac{1}{\pi}  \frac{\pi^3}{4}  =  \frac{\pi^2}{4}
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{ \frac{5 \pi^4}{16}}{( \frac{\pi^2}{4})^2} - 3 = 2
$$

---

$$p(a) \propto e^\frac{-a^2}{2} cosh(a)$$ 
The normalisation constant as calculated by scipy.integrate is $C = \frac{1}{\sqrt{2\pi e}}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{1}{\sqrt{2\pi e}} \int\limits_{-\infty}^{\infty} e^\frac{-a^2}{2} cosh(a) a^4 da = \frac{1}{\sqrt{2\pi e}} 10 \sqrt{2\pi e} = 10
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{1}{\sqrt{2\pi e}} \int\limits_{-\infty}^{\infty} e^\frac{-a^2}{2} cosh(a) a^2 da =  \frac{1}{\sqrt{2\pi e}}  2 \sqrt{2\pi e} =  2
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{10}{2^2} - 3 = -0.5
$$

---

$$p(a) \propto e^\frac{-a^4}{4}$$
The normalisation constant as calculated by scipy.integrate is $C = \frac{\sqrt{2}}{\Gamma(1/4)}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{\sqrt{2}}{\Gamma(1/4)} \int\limits_{-\infty}^{\infty} e^\frac{-a^4}{4} a^4 da = \frac{\sqrt{2}}{\Gamma(1/4)} \frac{\Gamma(1/4)}{\sqrt{2}} = 1
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{\sqrt{2}}{\Gamma(1/4)} \int\limits_{-\infty}^{\infty} e^\frac{-a^4}{4} a^2 da = \frac{\sqrt{2}}{\Gamma(1/4)} \Gamma(3/4)\sqrt{2} \approx  0.66
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{1}{0.66^2} - 3 \approx  -0.7
$$


---

$$p(a) \propto \frac{1}{(a^2 +5)^3}$$
The normalisation constant as calculated by scipy.integrate is $C = \frac{200\sqrt{5}}{3\pi}$
$$
\operatorname{E}[a^4] = \int\limits_{-\infty}^{\infty} p(a) a^4 da  = \frac{200\sqrt{5}}{3\pi} \int\limits_{-\infty}^{\infty}  \frac{1}{(a^2 +5)^3} a^4 da = \frac{200\sqrt{5}}{3\pi} \frac{3 \pi}{8\sqrt{5}} = 25
$$
$$
\operatorname{E}[a^2] = \int\limits_{-\infty}^{\infty} p(a) a^2 da  = \frac{200\sqrt{5}}{3\pi} \int\limits_{-\infty}^{\infty}  \frac{1}{(a^2 +5)^3} a^2 da = \frac{200\sqrt{5}}{3\pi} \frac{\pi}{40\sqrt{5}} = \frac{5}{3}
$$
$$
\verb+Kurt+[X] = \frac{\operatorname{E}[X^4]}{(\operatorname{E}[X^2])^2} - 3 = \frac{25}{(\frac{5}{3})^2} - 3 = 6
$$
