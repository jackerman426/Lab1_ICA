```python
#Functions needed
%pylab inline
# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
 return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp
def sine_wave(x, period=0.2, amp=1.0, phase=0.):
 return np.sin((x / period - phase) * 2 * np.pi) * amp
def square_wave(x, period=0.2, amp=1.0, phase=0.):
 return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) 
def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp
def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
# Generate sample data
num_sources = 5
signal_length = 500
t = linspace(0, 1, signal_length)
S = np.c_[sawtooth(t), sine_wave(t, 0.3), square_wave(t, 0.4), triangle_wave(t, 0.25), np.random.randn(t.size)].T

plot_signals(S)
```

### 1.4 Whitening

```python
def whiten(X):

   from numpy import dot, sqrt, diag
   from numpy.linalg import eigh

   # the matrix X should be observations-by-components

   # get the covariance matrix
   Xcov = np.cov(X)

   # eigenvalue decomposition of the covariance matrix
   d,V = eigh(Xcov)

   # a fudge factor can be used so that eigenvectors associated with
   # small eigenvalues do not get overamplified.
   D = diag(1./sqrt(d))

   # whitening matrix
   W = dot(dot(V,D),V.T)

   # multiply by the whitening matrix
   Xwhiten = dot(W,X)

   return Xwhiten

Xwhiten = whiten(X)
```

```python
from numpy import linalg as la
#activation functions

a0 = lambda x: -np.tanh(x);
a1 = lambda x: -x + tanh(x);
a2 = lambda x: -x**3;
a3 = lambda x: -((6*x)/(x**2 + 5));

def onlineICA(X, activation_function,iterations, lr=0.001):
    #Whiten the data
    X = whiten(X)
    #Create random unmixing matrix
    W = identity(len(X))
    for m in range(0,iterations):
        for x in X.T:
            a = dot(W,x)
            z = activation_function(a)
            xPrime = dot(W.T,a)
            #Wpinv = linalg.inv(W.T)
            dW = zeros((W.shape[0],W.shape[0]))
            for i in range(W.shape[0]):
                for j in range(W.shape[0]):
                    dW[i][j] = (W[i][j] + xPrime[j]*z[i])
            W = W + lr*dW
    anew = dot(W,X)
    return anew

def batchICA(X, activation_function, lr = 0.01 , thresh = 0.01):
    #Whiten the data
    X = whiten(X)
    W = random_nonsingular_matrix(len(X))
    count = 0;
    check = 2;
    while check > thresh:
        dW = zeros((W.shape[0],W.shape[0]))
        a = dot(W,X)
        z = activation_function(a)
        xPrime = dot(W.T,a)
        dW = W*X.shape[1] + dot(z,xPrime.T)
        W = W + lr*(dW/X.shape[1])
        check = sum(np.absolute(dW/X.shape[1]))
        count +=1
        if count%2 == 0 : print count, check
    anew = dot(W,X)
    return anew





#def ICA(X, activation_function, learning_rate):
```
