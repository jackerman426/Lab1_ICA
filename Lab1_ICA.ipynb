```python
#Functions needed
%pylab inline
# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
 return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp
def sine_wave(x, period=0.2, amp=1.0, phase=0.):
 return np.sin((x / period - phase) * 2 * np.pi) * amp
def square_wave(x, period=0.2, amp=1.0, phase=0.):
 return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) 
def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp
def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix if shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X):
    """
    Plot the signals contained in the rows of X.
    """
    figure()
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
```

```python
# Generate sample data
num_sources = 5
signal_length = 500
t = linspace(0, 1, signal_length)
S = np.c_[sawtooth(t), sine_wave(t, 0.3), square_wave(t, 0.4), triangle_wave(t, 0.25), np.random.randn(t.size)].T

plot_signals(S)
```

### 1.1 Make Mixtures

```python
def make_mixtures(S, A):
    X = np.dot(A,S)
    return X
#Get the Mixing Matrix
A = random_nonsingular_matrix(d=5)
#Get the mixed signals X
X = make_mixtures(S,A)
#Plot the mixed signals X
plot_signals(X)
```

### 1.3 Implicit priors

```python
from functools import partial
import math

def f0(a):
    y = []
    for i in a:
        y += [ 2*math.cosh(i)/(math.cosh(2*i) + 1) ]
    return y

def af0(a):
    y = []
    for i in a:
        y +=  [ -math.tanh(i)] 
    return y

def f1(a):
    y = []
    for i in a:
        y+= [ math.exp( -(i**2)/2 ) * math.cosh(i)]
    return y

def af1(a):
    y = []
    for i in a:
        y+= [ -i + math.tanh(i) ] 
    return y

def f2(a):
    y = []
    for i in a:
        y += [math.exp(-(i**4)/4)]
    return y

def af2(a):
    y = []
    for i in a:
        y += [-i**3]
    return y

def f3(a):
    y = []
    for i in a:
        y += [ 1/(i**2 + 5)**3 ]
    return y

def af3(a):
    y = []
    for i in a:
        y += [ -(6*i)/(i**2 + 5) ]
    return y
    
def plot_function(f):
    x_min=-3
    x_max=3
    x = linspace(x_min, x_max)
    y = f(x)
    plot(x,y)

    
figure(0)
subplot(1,2,1)
plot_function(af0)
subplot(1,2,2)
plot_function(f0)

figure(1)
subplot(1,2,1)
plot_function(af1)
subplot(1,2,2)
plot_function(f1)

figure(2)
subplot(1,2,1)
plot_function(af2)
subplot(1,2,2)
plot_function(f2)

figure(3)
subplot(1,2,1)
plot_function(af3)
subplot(1,2,2)
plot_function(f3)


```

### 1.4 Whitening

```python
def whiten(X):

   from numpy import dot, sqrt, diag
   from numpy.linalg import eigh

   # the matrix X should be observations-by-components

   # get the covariance matrix
   Xcov = np.cov(X)

   # eigenvalue decomposition of the covariance matrix
   d,V = eigh(Xcov)

   # a fudge factor can be used so that eigenvectors associated with
   # small eigenvalues do not get overamplified.
   D = diag(1./sqrt(d))

   # whitening matrix
   W = dot(dot(V,D),V.T)

   # multiply by the whitening matrix
   Xwhiten = dot(W,X)

   return Xwhiten

Xwhiten = whiten(X)
```

### 1.5 Interpret results of whitening 

```python
# scatter plots of the sources(blue), measurements(red) and whitened measurements(green)
def scatter_plot(S,X,Xwhiten):
    count = 1
    figure()
    for i in range (0,4):
        for j in range (i+1,5):
            subplot(10,3,count)
            plt.scatter(S[i],S[j],s=0.9,c='b')
            subplot(10,3,count+1)
            plt.scatter(X[i],X[j],s=0.9,c='r')
            subplot(10,3,count+2)
            plt.scatter(Xwhiten[i],Xwhiten[j],s=0.9,c='g')
            count += 3
rcParams['figure.figsize'] = 16, 40
scatter_plot(S,X,Xwhiten)
```
